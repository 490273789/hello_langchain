## PromptTemplate - 提示词模版
模版+变量值=完整的提示词 

### Prompt Template

```python
from langchain_core.prompts import PromptTemplate

# 1. 直接实例化Template
template = PromptTemplate(
    template="你是一个翻译助手，帮助用户将{content}翻译成语言：{lang}",
    input_variables=["content", "lang"],
)

# 调用方式1
format1 = template.format(content="你好", lang="法语")
print(f"format1: {format1}")
# template.invoke(format)

# 调用方式2 invoke + to_string
format2 = template.invoke({"content": "你好", "lang": "法语"})
print(f"format2.to_messages: {type(format2.to_messages())}")
print(f"format2: {format2.to_string()}")
print(f"format2.to_messages: {format2.to_messages()}")
```

```python
# 调用类方法构造提示词模版
template2 = PromptTemplate.from_template(
    template="你是一个翻译助手，帮助用户将{content}翻译成语言：{lang}",
)
template2
```
